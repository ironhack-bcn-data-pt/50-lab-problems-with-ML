## Developing a Model

### Implementation: Define a Performance Metric
What is the performance meteric with which you will determine the performance of your model? Create a function that calculates this performance metric, and then returns the score. 

```python
from sklearn.metrics import r2_score

def performance_metric(y_true, y_predict):
    """ Calculates and returns the performance score between 
        true and predicted values based on the metric chosen. """
    return r2_score(y_true, y_predict)

    
```

### Implementation: Shuffle and Split Data
Split the data into the testing and training datasets. Shuffle the data as well to remove any bias in selecting the traing and test. 

```python
#suffle the dataframe:
boston=boston.sample(frac=1).reset_index(drop=True)
```

```python
X=boston.drop(['MEDV'],axis=1)
y=boston['MEDV']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.2, random_state=5)
```

----

## Analyzing Model Performance
Next, we are going to build a Random Forest Regressor, and test its performance with several different parameter settings.

### Learning Curves
Lets build the different models. Set the max_depth parameter to 2, 4, 6, 8, and 10 respectively. 

```python
from sklearn import ensemble
from sklearn.ensemble import RandomForestRegressor
```

```python
regr1 = RandomForestRegressor(max_depth=2, random_state=0)
regr1.fit(X_train, y_train)
```

```python
regr2 = RandomForestRegressor(max_depth=4, random_state=0)
regr2.fit(X_train, y_train)
```

```python
regr3 = RandomForestRegressor(max_depth=6, random_state=0)
regr3.fit(X_train, y_train)
```

```python
regr4 = RandomForestRegressor(max_depth=8, random_state=0)
regr4.fit(X_train, y_train)
```

```python
regr5 = RandomForestRegressor(max_depth=10, random_state=0)
regr5.fit(X_train, y_train)
```

Now, plot the score for each tree on the training set and on the testing set.

```python
y_pred1=regr1.predict(X_test)
y_pred2=regr2.predict(X_test)
y_pred3=regr3.predict(X_test)
y_pred4=regr4.predict(X_test)
y_pred5=regr5.predict(X_test)
```

```python
r2_score(y_test,y_pred1)
```

```python
x1=[r2_score(y_test,y_pred1),
   r2_score(y_test,y_pred2),
   r2_score(y_test,y_pred3),
   r2_score(y_test,y_pred4),
   r2_score(y_test,y_pred5)]
```

```python
y_pred1=regr1.predict(X_train)
y_pred2=regr2.predict(X_train)
y_pred3=regr3.predict(X_train)
y_pred4=regr4.predict(X_train)
y_pred5=regr5.predict(X_train)

x2=[r2_score(y_train,y_pred1),
   r2_score(y_train,y_pred2),
   r2_score(y_train,y_pred3),
   r2_score(y_train,y_pred4),
   r2_score(y_train,y_pred5)]
```

```python
plt.plot([2,4,6,8,10],x1)
plt.plot([2,4,6,8,10],x2)
```

What do these results tell you about the effect of the depth of the trees on the performance of the model?

```python
#El depth óptimo es el 4 porque es más sencillo y tiene el score más alto. Si le vamos añadiendo más depth perdemos score y no lo recuperamos hasta 10 y el modelo se vuelve muy complejo
```

### Bias-Variance Tradeoff
When the model is trained with a maximum depth of 1, does the model suffer from high bias or from high variance? How about when the model is trained with a maximum depth of 10?

```python
regr = RandomForestRegressor(max_depth=1, random_state=0)
regr.fit(X_train, y_train)
y_pred=regr.predict(X_test)
r2_score(y_test,y_pred)
#The model suffer from high bias
#El modelo es muy simple y no se ha ajustado a los datos de entrenamiento (suele ser underfitting), por lo que produce un error alto en todas las muestras: entrenamiento, validación y test.
```

```python
regr5.fit(X_train, y_train)
y_pred5=regr5.predict(X_test)
r2_score(y_test,y_pred5)
#The model may suffer from high variance.
#Un modelo con varianza alta quiere decir que pequeños cambios en el dataset conlleva a grandes cambios en el output (suele ser overfitting)
```
